```python
from bokeh.plotting import figure, show, output_notebook, output_file
from bokeh.models import ColumnDataSource, Range1d
from bokeh.models.tools import HoverTool
from datetime import datetime
from bokeh.charts import Bar
output_notebook()
import pandas as pd
```

```python
projects = [
    ['Meta-analysis', 'Review', '2017-09-01', '2017-10-01'],
    ['Aerodigestive', 'Analysis', '2017-08-01', '2017-09-01'],
    ['Aerodigestive', 'Writing', '2017-09-01', '2017-10-15'],
    ['Aerodigestive', 'Review', '2017-12-01', '2018-02-01'],
    ['Metabolomics', 'Project development', '2017-08-01', '2017-12-01'],
    ['Metabolomics', 'Analysis', '2018-01-01', '2018-08-01'],
    ['Metabolomics', 'Writing', '2018-08-01', '2018-10-01'],
    ['Metabolomics', 'Review', '2018-12-01', '2019-01-01'],
    #['Underworlds', 'Analysis', '2017-11-01', '2018-01-01'],
    #['Underworlds', 'Review', '2018-03-01', '2018-05-01'],
    ['Annotrait', 'Project development', '2017-08-01', '2017-10-01'],
    ['Annotrait', 'Analysis', '2017-10-01', '2018-01-01'],
    ['Annotrait', 'Writing', '2018-01-01', '2018-03-01'],
    ['Annotrait', 'Review', '2018-05-01', '2018-07-01'],
    #['Thesis', 'Writing', '2018-10-01','2018-12-31']
]
projects = pd.DataFrame(projects, columns=['project', 'task', 'start', 'end'])
projects['start'] = pd.to_datetime(projects['start'])
projects['end'] = pd.to_datetime(projects['end'])
```

```python
colordict = {'Analysis': '#7030A0', 'Writing': '#FF6376', 'Review': '#C7C7C7', 'Project development': '#699BFF'}
#colordict = {'Analysis': '#C7C7C7', 'Writing': '#C7C7C7', 'Review': '#C7C7C7', 'Project development': '#C7C7C7'}

#colordict = {'Analysis': '#387852', 'Writing': '#FF6376', 'Review': '#DDDDDD', 'Project development': '#699BFF'}
#colordict = {'Analysis': '#DB5461', 'Writing': '#7A6F9B', 'Review': '#C7C7C7', 'Project development': '#699BFF'}
projects['color'] = projects['task'].apply(lambda x: colordict[x])

order = [
    #'Thesis', 
    #'Underworlds', 
    'Metabolomics', 
    'Annotrait', 
    'Aerodigestive', 
    'Meta-analysis'
]

projects['bottom'] = projects['project'].apply(lambda x: order.index(x) + 0.8)
projects['top'] = projects['project'].apply(lambda x: order.index(x) + 1.2)
projects
```

```python
p = figure(x_axis_type='datetime', width=800, height=300,
           x_range=Range1d(projects['start'].min(), projects['end'].max()),
           y_range=order)
CDS=ColumnDataSource(projects)
p.quad(left='start', right='end', bottom='bottom', top='top', source=CDS, color='color')
p.grid[0].ticker.desired_num_ticks = 20
p.legend.location = "top_right"
```

```python
show(p)
```
